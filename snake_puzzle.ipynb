```python
def snake(numbers):
    # nums to str
    nums = [str(n) for n in numbers]

    # first num as start
    spine = [nums[0]]
    remaining = nums[1:]

    while remaining:
        current_last_two = spine[-1][-2:]
        found = False
        
        for i, next_num in enumerate(remaining):
            if current_last_two == next_num[:2]:
                spine.append(remaining.pop(i))
                found = True
                break
        
        if not found:
            break

    # Connection without doubles
    result = spine[0]
    for i in range(1, len(spine)):
        result += spine[i][2:] 
        
    return result

# test
#numbers = [127856, 564321, 413278, 542710, 212754]
#print(build_chain(numbers))
```
